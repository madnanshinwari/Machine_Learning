```python
# fill Age column add those values which is missing 
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
# apply label Encoding on Gender column 
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender']) # Encode column

print(df.head())
```
